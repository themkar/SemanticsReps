```python
from google.colab import auth
import gspread
from google.auth import default
#autenticating to google
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
import pandas as pd
import numpy as np
#defining my worksheet
worksheet = gc.open('Quotes').sheet1
#get_all_values gives a list of rows
rows = worksheet.get_all_values()
#Convert to a DataFrame 
df = pd.DataFrame(rows)
df.columns = df.iloc[0]
df = df[1:]

# convert to numbers
df[['Warmth_C1','Warmth_C2','Warmth_C3','Warmth_C4','Warmth_C5']] = df[['Warmth_C1','Warmth_C2','Warmth_C3','Warmth_C4','Warmth_C5']].apply(pd.to_numeric, errors='coerce')
df[['WarmthConfidence_C1','WarmthConfidence_C2','WarmthConfidence_C3','WarmthConfidence_C4','WarmthConfidence_C5']] = df[['WarmthConfidence_C1','WarmthConfidence_C2','WarmthConfidence_C3','WarmthConfidence_C4','WarmthConfidence_C5']].apply(pd.to_numeric, errors='coerce')
df['WARMTH'] =df[['Warmth_C1','Warmth_C2','Warmth_C3','Warmth_C4','Warmth_C5']].mean(axis = 1 )  
df['WARMTHConfidence'] =df[['WarmthConfidence_C1','WarmthConfidence_C2','WarmthConfidence_C3','WarmthConfidence_C4','WarmthConfidence_C5']].mean(axis = 1 )  
df[['Competence_C1','Competence_C2','Competence_C3','Competence_C4','Competence_C5']] = df[['Competence_C1','Competence_C2','Competence_C3','Competence_C4','Competence_C5']].apply(pd.to_numeric, errors='coerce')
df[['CompetenceConfidence_C1','CompetenceConfidence_C2','CompetenceConfidence_C3','CompetenceConfidence_C4','CompetenceConfidence_C5']] = df[['CompetenceConfidence_C1','CompetenceConfidence_C2','CompetenceConfidence_C3','CompetenceConfidence_C4','CompetenceConfidence_C5']].apply(pd.to_numeric, errors='coerce')
df['COMPETENCE'] =df[['Competence_C1','Competence_C2','Competence_C3','Competence_C4','Competence_C5']].mean(axis = 1 )  
df['COMPETENCEConfidence'] =df[['CompetenceConfidence_C1','CompetenceConfidence_C2','CompetenceConfidence_C3','CompetenceConfidence_C4','CompetenceConfidence_C5']].mean(axis = 1 )  
df['WARMTHConfidence']=df['WARMTHConfidence']/11
df['COMPETENCEConfidence']=df['COMPETENCEConfidence']/11

df['WARMTHCOMPETENCE']=df['WARMTH']*df['WARMTH']+df['COMPETENCE']*df['COMPETENCE'] 

df[['Year','Month','VADER','FLAIR','TextBlob']] = df[['Year','Month','VADER','FLAIR','TextBlob']].apply(pd.to_numeric, errors='coerce')


df['size_ratio']= 1

df['WARMTH_codable']= 5-1* df.Warmth_C1.isna()-1* df.Warmth_C2.isna()-1* df.Warmth_C3.isna()-1* df.Warmth_C4.isna()-1* df.Warmth_C5.isna()
df['COMPETENCE_codable']= 5-1* df.Competence_C1.isna()-1* df.Competence_C2.isna()-1* df.Competence_C3.isna()-1* df.Competence_C4.isna()-1* df.Competence_C5.isna()
df['WARMTH_COMPETENCE_codable'] = 1*(df['WARMTH_codable']>=3)*(df['COMPETENCE_codable']>=3) 

df['isBROADSHEET']=(df['Newspaper']=='Telegraph') + (df['Newspaper']=='Observer')+(df['Newspaper']=='Independent')+(df['Newspaper']=='Guardian')+(df['Newspaper']=='Times')
df['isLEFT']= (df['Newspaper']=='Observer')+(df['Newspaper']=='Independent')+(df['Newspaper']=='Guardian')+(df['Newspaper']=='Mirror')


df['QUARTILE'] = 1 + 1*(df['WARMTH']>=0.5) + 2*(df['COMPETENCE']>=0.5) 
df.loc[df['QUARTILE'] == 1, 'QUARTILE'] = 'Low Warmth, Low Competence'
df.loc[df['QUARTILE'] == 2, 'QUARTILE'] = 'High Warmth, Low Competence'
df.loc[df['QUARTILE'] == 3, 'QUARTILE'] = 'Low Warmth, High Competence'
df.loc[df['QUARTILE'] == 4, 'QUARTILE'] = 'High Warmth, High Competence'


df = df.round(3)
df.head(30)
```

```python
import plotly.express as px 
import numpy as np

fig = px.scatter(data_frame = df,
    x = df.WARMTH,
    y = df.COMPETENCE,
    size = df.size_ratio,
    size_max = 10,
    title = 'RATINGS - ALL 1000 QUOTES - color shows CONFIDENCE',
    trendline="ols",
    color = df.WARMTHConfidence+df.COMPETENCEConfidence,
    hover_name = 'Quote',
    hover_data = ['ID','Newspaper','Year','WARMTHConfidence','COMPETENCEConfidence','WARMTH_codable', 'COMPETENCE_codable'])

xlim = [-0.1, 1.1]
ylim = [-0.1, 1.1]

fig.update_layout(
    height = 1000,
    width = 3000,
    margin=dict(l=50, r=2000, t=50, b=50),
    paper_bgcolor="LightSteelBlue",
)

fig.show()
```
